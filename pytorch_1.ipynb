```python
import numpy as np
import matplotlib.pyplot as plt

# True parameters (unknown in real-world scenarios)
true_mu = 5  # True mean
true_sigma = 2  # True standard deviation

# Generate sample data
np.random.seed(42)
n_samples = 20  # Small dataset to show MAP advantage
samples = np.random.normal(true_mu, true_sigma, n_samples)

# MLE Estimate (Sample Mean)
mle_mu = np.mean(samples)

# MAP Estimate (Assume Gaussian Prior: mu ~ N(mu_0, tau^2))
mu_0 = 3   # Prior mean (our belief about mu)
tau_sq = 4  # Prior variance (uncertainty in prior belief)

# Compute MAP estimate using Bayesian formula
map_mu = (n_samples * mle_mu + mu_0 / tau_sq) / (n_samples + 1 / tau_sq)

# Print results
print(f"True Mean: {true_mu}")
print(f"MLE Estimate: {mle_mu:.4f}")
print(f"MAP Estimate: {map_mu:.4f} (incorporating prior belief)")

# Plot results
plt.figure(figsize=(8, 5))
plt.axvline(true_mu, color='r', linestyle='--', label='True Mean')
plt.axvline(mle_mu, color='b', linestyle='-', label='MLE Estimate')
plt.axvline(map_mu, color='g', linestyle='-.', label='MAP Estimate')
plt.legend()
plt.title("MLE vs MAP Estimation of Mean")
plt.xlabel("Mean Value")
plt.show()
```
